#Part 1: Analysis of Undervalued Companies — PEG Ratio vs. P/FCF Ratio

At the beginning of our journey, we explored finviz.com to collect lots of public information. We put all this important metadata into one neat CSV file. But our path is not without its challenges; the raw information must be cleaned and refined. This process is called an ETL pipeline.

In this part, we explore the correlation between the Price/Earnings-to-Growth (PEG) ratio and the Price-to-Free Cash Flow (P/FCF) ratio among the 8549 companies within our dataset, all of which are part of the US stock market.

The Price/Earnings to Growth (PEG) ratio is a valuation metric that is often used to assess whether a stock is overvalued or undervalued based on both its price-to-earnings (P/E) ratio and its expected earnings growth rate. Here’s why the PEG ratio is considered important in identifying undervalued stocks:

Incorporates Earnings Growth: The PEG ratio takes into account not only a company’s current earnings (P/E ratio) but also its expected earnings growth rate. This is important because a stock with a low P/E ratio might seem cheap, but if its earnings are expected to grow significantly in the future, it could still be undervalued.
Relative Valuation: The PEG ratio provides a more normalized valuation measure compared to just the P/E ratio. It helps investors compare the price of a stock to its growth prospects, making it easier to evaluate companies with different growth rates.
Identifying Undervalued Stocks: A PEG ratio less than 1 is often interpreted as a signal that a stock may be undervalued. This suggests that the market’s valuation of the stock is not fully reflecting its expected earnings growth. However, it’s important to note that a low PEG ratio does not guarantee that a stock is undervalued; it simply suggests that the stock might warrant further investigation.
Balancing Risk and Return: The PEG ratio attempts to balance the trade-off between the potential for high growth and the risk associated with investing in a high-growth company. A low PEG ratio can indicate that a stock might be a relatively attractive investment in terms of potential return compared to the level of risk.
Long-Term Perspective: The PEG ratio encourages investors to consider a longer-term outlook by factoring in earnings growth. This can be particularly relevant when evaluating stocks in industries or sectors that are expected to experience significant growth over time.
The Price/Free Cash Flow (P/FCF) ratio is a financial metric used to evaluate the valuation of a company’s stock relative to its free cash flow. Free cash flow represents the cash a company generates from its operations after accounting for capital expenditures necessary to maintain and expand its business. The P/FCF ratio is similar to the more common Price/Earnings (P/E) ratio but uses free cash flow instead of earnings.

In summary, the P/FCF ratio provides insights into the company’s financial health, cash flow sustainability, and potential valuation relative to its peers. However, like the other factors it should be used as part of a comprehensive analysis that considers other financial and qualitative factors.

We have refined the dataset by filtering out companies with a market capitalization of over 1 billion dollars and focusing on specific industry sectors, namely excluding Financial, Healthcare, Real Estate, Communication Service, and Technology. Additionally, we have set the criteria that P/FCF should be below 200, and PEG should be below 1 while ensuring that both the SMA200 and SMA50 indicators are greater than 0. SMA50 (Simple Moving Average 50) and SMA200 (Simple Moving Average 200) are used to assess the trend and potential value of a stock.
